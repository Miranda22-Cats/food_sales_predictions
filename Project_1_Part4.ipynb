```python
#import libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
sales_predictions = '/content/sales_predictions.csv'
df= pd.read_csv('sales_predictions.csv')
df.head()
```

```python
df.info()
```

Does Item_MRP increase the Item_Outlet_Sales?



```python
sns.scatterplot(data=df, x = "Item_MRP", y= 'Item_Outlet_Sales');
plt.title('Effect of Product MRP on Outlet Sales', fontsize = 14);
```

```python
sns.regplot(data=df, x= 'Item_MRP', y= 'Item_Outlet_Sales', line_kws={'color':'black', 'ls': '--'}, scatter_kws={'s':10});
plt.title('Effect of Product MRP on Outlet Sales', fontsize = 14);
```

Seems to be a trend that as the Item_MRP increases, Item_Outlet_Sales increases.

Does the Item_Visibility effect the Item_Outlet_Sales?

```python
sns.scatterplot(data=df, x='Item_Visibility', y= 'Item_Outlet_Sales');
sns.regplot(data=df, x='Item_Visibility', y= 'Item_Outlet_Sales', line_kws={'color':'black', 'ls': '--'}, scatter_kws={'s':10});
plt.title('Effect of Item Visibility on Outlet Sales', fontsize = 14);
```

Seems to be a negative correlation between the Item Visibility and Item Outlet Sales. AS the Items Visibility increases, Item Outlet Sales decrease, which is strange.
